```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/content/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

##Data Cleaning

```python
df.isnull().sum()
```

```python
df.drop("customerID", axis =1, inplace = True)
```

```python
df.info()
```

```python
df.shape
```

##Visualization

```python
sns.countplot(x='Churn', data = df)
plt.title("Customer Churn Distribution")
plt.show()
```

```python
#To check churn values Numerically
df['Churn'].value_counts()
```

```python
df.info()
```

```python
# To Convert TotalCharges into Numerical
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#To remove missing rows
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

##Data Preprocessing

```python
#To Covert Churn values from text to numeric
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
df.info()
```

```python
df['Churn'].head()
```

```python
#Converted all text columns into Numbers
df = pd.get_dummies(df, drop_first=True)
```

```python
df.info()
```

```python
df.shape
```

```python
x = df.drop('Churn', axis = 1)
y = df['Churn']
```

```python
x.shape
```

```python
y.shape
```

##Training and Testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

##Model Building

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)

#Prediction
y_pred = model.predict(x_test)
```

##Accuracy Score

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')

plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import classification_report
classification_report = classification_report(y_test, y_pred)
print(classification_report)
```

##Conclusion

###
A Customer Churn Prediction model was developed using the IBM Telco Customer Churn dataset. After preprocessing and encoding the data, a Logistic Regression model was trained and evaluated. The model achieved an accuracy of 78.75%, indicating good predictive performance. The analysis showed that factors such as tenure, contract type, and monthly charges influence customer churn. This model can help telecom companies identify customers who are likely to leave and take preventive retention actions.
